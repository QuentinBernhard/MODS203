```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
doctors = pd.read_csv('doctors.csv')
doctors.pop(doctors.columns[0])
doctors.pop(doctors.columns[4])
doctors

def convert_time(date):
    if date == "pas de disponibilités":
        return 0
    else:
        from datetime import datetime, timedelta
        date1 = date[0] + date[1]
        print(date1)
        date2 = "jeu. 5 janv."

        # Convert the string dates to datetime objects
        try:
            date1 = datetime.strptime(date1, "%A. %d %b.")
            date2 = datetime.strptime(date2, "%A. %d %b.")

            # Calculate the difference between the two dates
            difference = (date1 - date2)
            print(difference)
            difference= difference.total_seconds()/3600
            

            # Print the difference as a timedelta object
            print(difference)
        except:
            pass


doctors['Waiting Time'] = doctors['Waiting Time'].apply(convert_time)
doctors
doctors.dropna()

plt.hist(doctors['Waiting Time'])
```
